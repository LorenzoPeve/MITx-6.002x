```python
import matplotlib.pyplot as plt
```

```python
mySamples = []
myLinear = []
myQuadratic = []
myCubic = []
myExponential = []
```

```python
for i in range(0, 30):
    mySamples.append(i)
    myLinear.append(i)
    myQuadratic.append(i**2)
    myCubic.append(i**3)
    myExponential.append(1.5**i)
```

```python
fig, ax = plt.subplots()
ax.plot(mySamples, myLinear, linewidth = 2, label = '$x$')
ax.plot(mySamples, myQuadratic, linewidth = 2, label = '$x^2$')
ax.plot(mySamples, myCubic, linewidth = 2, label = '$x^3$')
ax.plot(mySamples, myExponential, linewidth = 2, label = '$1.5^x$')
plt.title('Visualizing Algorithm Complexity')
plt.legend()
plt.show()
```

## Compound Interest

Planning for retirement:
* intend to save an amount `m` each month
* expect to earn a percentage `r` of income on investments each month

How big will be the retirement fund?

Impact of Contribution

```python
def retire(monthly: float, rate: float, terms: int):

    """
    ArgS:
        monthly (float): Amount contributed monthly
        rate (float): Annual return rate
        terms (int): Number of months
    
    """
    savings = [0]
    base = [0]
    mRate = rate/12
    for i in range(terms):
        base += [i]
        savings += [savings[-1]*(1 + mRate) + monthly]

    return base, savings 
```

```python
def displayRetireWMonthlies(monthlies, rate, terms):
    fig, ax = plt.subplots()

    for monthly in monthlies:
        xvals, yvals = retire(monthly, rate, terms)
        ax.plot(xvals, yvals, label = 'retire:'+str(monthly))
    
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles[::-1], labels[::-1], loc='upper left')
    plt.show()
    
years = 40
displayRetireWMonthlies([500, 600, 700, 800, 900, 1000, 1100], .05, years * 12)
```

Impact of Rate of Return

```python
def displayRetireWRates(month, rates, terms):
    fig, ax = plt.subplots()

    for rate in rates:
        xvals, yvals = retire(month, rate, terms)
        ax.plot(xvals, yvals, label = 'retire:'+str(month)+ ':' + \
            str(int(rate*100)))

    
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles[::-1], labels[::-1], loc='upper left')
    plt.show()
    
years = 40
displayRetireWRates(800, [0.03, 0.05, 0.07], years * 12)
```
